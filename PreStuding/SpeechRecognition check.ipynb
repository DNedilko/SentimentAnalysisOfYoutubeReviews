```python
# import SR library
import speech_recognition as sr
from AudioSplit import SplitWavAudioMubin
from pydub import AudioSegment

def match_target_amplitude(sound, target_dBFS):
    change_in_dBFS = target_dBFS - sound.dBFS
    return sound.apply_gain(change_in_dBFS)

# crate an instance of recogniser
recognizer = sr.Recognizer()
# energy threshold - vals below are sounds, above - speech
recognizer.energy_threshold=300
# spaCy for extracting named entities and fillers also can be used for breaking on sentences
folder = "./wav_data_reviews"
file = ["1.wav","2.wav"]

split_wav = SplitWavAudioMubin(folder,file[0])
split_wav.multiple_split(min_per_split=3)
# clean_uk_speech = sr.AudioFile("./wav_data_reviews/2.wav")
file_list = next(os.walk("./wav_data_reviews"), (None, None, []))[2]
print(file_list)
full_eng_text = []
for f in file_list:
    if f == "1.wav" or f == "2.wav": continue

    sound = AudioSegment.from_file(f'{folder}/{f}', "wav")
    normalized_sound = match_target_amplitude(sound, -20.0)
    normalized_sound.export(f'./wav_normalized/norm_{f}', format="wav")
    clean_eng_speech = sr.AudioFile(f'./wav_normalized/norm_{f}')

    with clean_eng_speech as data:
        eng_audio = recognizer.record(data)
    eng_text = recognizer.recognize_google(audio_data=eng_audio,language="en-US",show_all=False)
    print(f'file {f} recognized')
    full_eng_text.append(eng_text)

# with clean_uk_speech as data:
#     uk_audio = recognizer.record(data)



# uk_text = recognizer.recognize_google(audio_data=uk_audio,language="uk",show_all=False)

full_text = " ".join(full_eng_text)
```

```python
# NLP part of the project
from deepmultilingualpunctuation import PunctuationModel
from nltk import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer

def stem(data, stemmer):
    stem_data=[]
    for sent in data:
        stem_temp = []
        for word in sent:
            stem_temp.append(stemmer.stem(word))
        stem_data.append(stem_temp)
    return stem_data

def data_to_tokens(doc):
    punctuations="?:!.,;"
    sentences = sent_tokenize(doc)
    sent_words = []
    for sentence in sentences:
        print(sentence, '\n')
        words = word_tokenize(sentence)
        for word in words:
            if word in punctuations: words.remove(word)
        sent_words.append(words)

    return sent_words


# porter = PorterStemmer(use_fast = False)
wordnet_lemmatizer = WordNetLemmatizer()

snow_stemmer = SnowballStemmer(language='english')
porter = PorterStemmer()
model = PunctuationModel()
result = model.restore_punctuation(full_text)
# lancaster=LancasterStemmer()


sent_words = data_to_tokens(full_text)
print(sent_words)
# stemmed_data = stem(sent_words, snow_stemmer)
#
# print(stemmed_data[1:3])
```
